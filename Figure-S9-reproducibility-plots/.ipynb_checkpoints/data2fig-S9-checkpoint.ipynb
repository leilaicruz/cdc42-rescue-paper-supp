```python
import scipy.io
from scipy import stats, optimize, interpolate
import pandas as pd
import numpy as np
from collections import defaultdict
import math
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

## General Data all strains

```python
names=['bem1d-all-experiments-8h','bem1dbem3d-all-experiments-8h','WT_CDC42-all-experiments-8h','WT-all-experiments-8h']

variable_names=['num028','num04','num06','num128','num14','num16','num228','num24','num26']

meaning=['0% Gal-28112018','0% Gal-04122018','0% Gal-06122018','0.06% Gal-28112018','0.06% Gal-04122018',
        '0.06% Gal-06122018','0.1% Gal-28112018','0.1% Gal-04122018','0.1% Gal-06122018']
strains=['bem1d','bem1dbem3d','WT+Cdc42','WT']
reads=defaultdict(dict)
```

## Reproducibility across experiments (fig S9) , case of bem1d strain

```python

r= scipy.io.loadmat(names[0] + '.mat')
for i in range(len(variable_names)): 
    reads[variable_names[i]]['perimeter-values']=r[variable_names[i]]
    reads[variable_names[i]]['Names']=strains[0]+meaning[i]
    reads[variable_names[i]]['radii-values']=reads[variable_names[i]]['perimeter-values']/(2*np.pi)

    
df_bem1d=pd.DataFrame(reads)

```

## Reproducibility across experiments (fig S9) , case of bem1dbem3d strain

```python
r= scipy.io.loadmat(names[1] + '.mat')
for i in range(len(variable_names)): 
    reads[variable_names[i]]['perimeter-values']=r[variable_names[i]]
    reads[variable_names[i]]['Names']=strains[1]+meaning[i]
    reads[variable_names[i]]['radii-values']=reads[variable_names[i]]['perimeter-values']/(2*np.pi)

    
df_bem1dbem3d=pd.DataFrame(reads)

df_bem1dbem3d.T
```

## Reproducibility across experiments (fig S9) , case of WT+cdc42 strain

```python
r= scipy.io.loadmat(names[2] + '.mat')
for i in range(len(variable_names)): 
    reads[variable_names[i]]['perimeter-values']=r[variable_names[i]]
    reads[variable_names[i]]['Names']=strains[2]+meaning[i]
    reads[variable_names[i]]['radii-values']=reads[variable_names[i]]['perimeter-values']/(2*np.pi)

    
df_wt_cdc42=pd.DataFrame(reads)
df_wt_cdc42.T
```

## Reproducibility across experiments (fig S9) , case of WT strain

```python
variable_names_wt=['num04','num06','num14','num16','num24','num26']
meaning_wt=['0% Gal-04122018','0% Gal-06122018','0.06% Gal-04122018',
        '0.06% Gal-06122018','0.1% Gal-04122018','0.1% Gal-06122018']
r= scipy.io.loadmat(names[3] + '.mat')
for i in range(len(variable_names_wt)): 
    reads[variable_names_wt[i]]['perimeter-values']=r[variable_names_wt[i]]
    reads[variable_names_wt[i]]['Names']=strains[3]+meaning_wt[i]
    reads[variable_names_wt[i]]['radii-values']=reads[variable_names_wt[i]]['perimeter-values']/(2*np.pi)

    
df_wt=pd.DataFrame(reads)
df_wt.T
```

## All data is analyzed at once, each strain data set is analyzed independently

```python
#all data
keys=['bem1d', 'bem1bem3d','wt_cdc42','wt']
frames = [df_bem1d,df_bem1dbem3d,df_wt_cdc42,df_wt]
df_keys = pd.concat(frames, keys=keys,sort=False).T
df_keys.to_excel('data2figS9.xlsx')

#issue: I can not load the excel file into the same dataframe as the original df_keys
```

```python
pd_reads=df_keys
fig, axes = plt.subplots(4,1, figsize=(10,7), dpi=100, sharex=True, sharey=True,squeeze=False)
axis_font = {'fontname':'Arial', 'size':'12'}
plt.subplots_adjust(bottom=0, right=0.8, top=1.6)

i=0
colors = cm.rainbow(np.linspace(0, 1, len(df_keys[keys[0]].index)))
for y, c in zip(keys, colors):
    plt.subplot(4,1,i+1)
    plt.hist(pd_reads[y]['radii-values'],bins=20,density=True,label=pd_reads[y]['Names'],stacked=True,alpha=0.8,color=colors)
    i=i+1
    plt.legend(prop={'size': 10},loc='upper right')
    plt.grid(color='black', linestyle='--', linewidth=0.1)
    plt.xlabel('Cell radii')
    plt.ylabel('Normalized frequency')
    plt.ylim(0,1.3)
    plt.xlim(1,11)

```
