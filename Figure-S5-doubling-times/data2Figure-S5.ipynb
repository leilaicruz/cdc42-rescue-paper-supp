```python
import scipy.io
from scipy import stats, optimize, interpolate
import pandas as pd
import numpy as np
from collections import defaultdict
import math
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
names=['data_doublings','std_doublings']
variable_names=names
meaning=['bem1d-8h','bem1d-24h','bem1dbem3d-8h','bem1dbem3d-24h','WT+CDC42-8h','WT+CDC42-24h']

reads=defaultdict(dict)
r= scipy.io.loadmat(variable_names[0] + '.mat')
r_std= scipy.io.loadmat(variable_names[1] + '.mat')
a=r[names[0]].shape # number of columns of the data
for i in range (a[1]):
    reads[i]['Values'] =r[names[0]][:, i]
    reads[i]['Names'] =meaning[i]
    reads[i]['Std']=r_std[names[1]][:,i]

pd_reads=pd.DataFrame(reads).T
```

```python
pd_reads
```

```python
fig, axes = plt.subplots(3,2, figsize=(10,5), dpi=100, sharex=True, sharey=True)
kwargs = dict(alpha=0.5, bins=20, density=True, stacked=True)
plt.subplots_adjust(bottom=0.2, right=1.5, top=2)
axis_font = {'fontname':'Arial', 'size':'8'}
colors = cm.Blues(np.linspace(0.2, 1.2, 3))
a=r[names[0]].shape # number of columns of the data

for i in  range (a[1]):
    plt.subplot(3,2, i+1)
    plt.bar(x=['0% Gal','0.06% Gal','0.1% Gal'],height=pd_reads['Values'][i]/60,yerr=pd_reads['Std'][i]/60,align='center', alpha=0.5, ecolor='black', color=colors,capsize=10,label=({'0%'},{'0.06%'},{'0.1%'}))
    plt.ylabel('Doubling Times (hours)',**axis_font)
    plt.title(pd_reads['Names'][i])
    plt.ylim(0,20)
plt.tight_layout()
plt.savefig("Figure-doublings.png",dpi=300,format='png',transparent=True)
```
